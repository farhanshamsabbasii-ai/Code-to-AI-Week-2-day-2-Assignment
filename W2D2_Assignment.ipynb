Tum ek **mini database** banao apne kisi pasandeeda topic par — e.g. friends, books, players, cars, shops. Yeh ek **list of dictionaries** hogi. Phir us par operations karo.


## Step 1 — One dictionary
Apne topic ka ek item ek dictionary mein banao (kam se kam 3 keys).
Example: `{"name": "Babar Azam", "role": "Batsman", "matches": 100}`

````python
# TODO: ek dictionary banao
item = {"name" : "Farhan" , "age" : 19 , "course" : "Code to AI"} 

print(item)
````

## Step 2 — Access values
Apni dictionary se 2 values key ke zariye print karo.
Ek `.get()` se bhi print karo (safe access).

````python
# TODO: print 2 values by key
print(item["name"])
print(item["course"])

# TODO: print one value using .get()
print(item.get("age"))
````

## Step 3 — Build a list of dictionaries ⭐
Kam se kam **4 items** ki list banao (har ek dictionary).
Same keys use karo har item mein (taaki yeh ek table jaisa lage).

````python
# TODO: list of dictionaries (4+ items)
database = [
    {"name" : "Ali" , "age" : 27 , "favorite_game" : "Cricket", "lucky_number" : 24},
    {"name" : "Mahmood" , "age" : 23 , "favorite_game" : "Cricket", "lucky_number" : 14},
    {"name" : "Faisal" , "age" : 21 , "favorite_game" : "Chess", "lucky_number" : 27},
    {"name" : "Farhan" , "age" : 19 , "favorite_game" : "Chess", "lucky_number" : 136}
]

print("Total items:", len(database))
````

## Step 4 — Loop and print nicely
`for` loop chala kar har item ko ek line par print karo, achhe format mein.
Example output style:
```
Babar Azam | Batsman | 100 matches
```

````python
# TODO: loop over database and print each item nicely
for d in database:
    
    print("Name:" , d.get("name") , end=" | ")
    print("Age:" , d.get("age") , end=" | ")
    print("Favorite game:" , d.get("favorite_game") , end=" | ")
    print("Lucky number:" , d.get("lucky_number"))
````

## Step 5 — Add and update
1. Ek naya item `append()` se add karo
2. Kisi ek item ki ek value update karo (e.g. `database[0]["matches"] = 105`)
3. Updated list ko dobara loop karke print karo

````python
# TODO: append a new item
database.append({"name" : "Asim" , "age" : 22 , "favorite_game" : "Cricket", "lucky_number" : 94})
# TODO: update one value in one item
database[1]["favorite_game"] = "Table Tenis"

# TODO: loop and print again
for d in database:
    print("Name:" , d.get("name") , end=" | ")
    print("Age:" , d.get("age") , end=" | ")
    print("Favorite game:" , d.get("favorite_game") , end=" | ")
    print("Lucky number:" , d.get("lucky_number"))
````

## Step 6 — Sets: find unique values
Apni database se kisi ek field ki saari values ek list mein nikaalo,
phir `set()` se unique values print karo.
Example: saare "city" ya "role" — unique kitne hain.

````python
# TODO: collect one field's values into a list using a loop
values = []
for d in database:
    values.append(d["favorite_game"])
# for row in database:
#     values.append(row["role"])   # apni key use karo

values = set(values)

# TODO: print unique values with set()
print(values)
````

## 🌟 Bonus (optional)
`.items()` use karke apni list ke **pehle** item (dictionary) ke saare key-value pairs
ek loop mein print karo — `key -> value` format mein.

````python
# TODO: loop over database[0].items() and print "key -> value"

for key , value in database[0].items():
    print(key , "->", value)
````
